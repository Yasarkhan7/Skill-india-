# **2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
data=[]
ls=0
for Sport in df['Sport'].unique():
    for i in range (0,len(df)):
        if (df[['Sport']].values[i]==Sport and df[['Medal']].values[i]=='Gold'):
            ls+=1
    data.append([Sport,ls])  
    print(Sport,ls)      
    ls=0
print(data)
    
```

```python
pd.DataFrame(data,columns = ['Sport','ls']).sort_values(by='ls', ascending=False).head().plot(x = 'Sport', y = 'ls', kind = 'bar', figsize = (10,10))
```

# **3. Which sport is having most number of medals so far? (Top 5)**

```python
data=[]
Medal=0
for Sport in df['Sport'].unique():
    for i in range (0,len(df)):
        if (df[['Sport']].values[i]==Sport and (df[['Medal']].values[i]=='Gold' or df[['Medal']].values[i]== 'Bronze' or df[['Medal']].values[i]== 'Silver' )):
            Medal+=1
    data.append([Sport,Medal])       
    Medal=0
print(data)
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (10,10))
```

# **4. Which player has won most number of medals? (Top 5)**

```python
data=[]
Medal=0
for Athlete in df['Athlete'].unique():
    for i in range (0,len(df['Athlete'])):
        if (df['Athlete'].values[i]==Athlete and (df['Medal'].values[i]=='Gold' or df['Medal'].values[i]== 'Bronze' or df['Medal'].values[i]== 'Silver' )):
            Medal+=1
    data.append([Athlete,Medal])       
    Medal=0

```

```python
pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (10,10))
```
